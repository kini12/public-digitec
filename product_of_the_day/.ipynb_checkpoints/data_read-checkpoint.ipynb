```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import datetime as dt
```

```python
df = pd.read_csv("Digitec.csv", encoding='latin-1', sep=",")
```

```python
pd.options.display.float_format = "{:,.2f}".format
```

```python
df.head(60)
```

```python
discount = df.old_price - df.new_price
```

```python
# old price - new price = discount
discount = discount.round(2)
```

```python
# 1%
max = df.old_price/100 
```

```python
discount_pourcentage = discount / max
```

```python
discount_pourcentage.round(2)
```

```python
df["discount"]= discount
```

```python
df["%"]=discount_pourcentage.round(2)
```

```python
data1 = df["%"]
data2 = df["discount"]
Date = df["date"]
```

```python
df.mean().round(2)
```

```python
fig=plt.figure(figsize=(80,40))
plt.plot(Date, data1, color='#444444',
         linestyle='-', label='%', marker = ".")

plt.plot(Date, data2, label='discount prize', linewidth=1, marker= "x")

plt.yticks(rotation=45)
plt.xticks(rotation=35) # en bas

plt.grid(True,which="both", linestyle='--')

plt.tick_params(axis = "x", labelsize = 6)
plt.xticks(Date[::10])

#plt.xscale("linear")

# acces a part of the chart
#plt.xlim(20,40)
#plt.ylim(0,100)

plt.title("Digitec discount")

plt.legend(loc = 'upper left')


plt.show()
plt.close()
```

```python
#creation of df with % as index, from biggest to lowest
df.sort_values(by = "%", ascending = False).head(50)
```

```python
#mean of %
moy = df["%"].mean()
moy = str(round(moy, 2))
moy
```
